# lets return all departments ,which were active all the 6 months 

```python

nn=[]
bb=[]
dd=[]
ii=[]
for index, row in messages.iterrows():
    for i, r in employees.iterrows():
        if (row['receiver']==r['id']):
            dd.append(r['department'])
            bb.append(r['location'])
            nn.append(r['age'])
            ii.append(r['id'])
            
        if (row['sender']==r['id']):
            dd.append(r['department'])
            bb.append(r['location'])
            nn.append(r['age'])
            ii.append(r['id'])
            
```

```python
aa=[]

for index, row in messages.iterrows():
    for i, r in employees.iterrows():
        if (row['receiver']==r['id']):
            aa.append(r['department'])

            
```

```python
cc=[]
for index, row in messages.iterrows():
    for i, r in employees.iterrows():
        if (row['sender']==r['id']):
            cc.append(r['department'])
            
```

```python
Dep_R=pd.DataFrame(aa,columns=['department'])
print(Dep_R.head(5))
Dep_R.info()
Dep_S=pd.DataFrame(cc,columns=['department'])
print(Dep_S.head(5))
Dep_S.info()
```

```python
Dep=pd.DataFrame(dd,columns=['department'])
print(Dep.head(5))
Dep.info()
Dep_L=pd.DataFrame(bb,columns=['location'])
print(Dep_L.head(5))
Dep_L.info()
Dep_A=pd.DataFrame(nn,columns=['age'])
print(Dep_A.head(5))
Dep_A.info()
```
